```python
from astropy.coordinates import SkyCoord
from astropy import units as u
from astroquery.simbad import Simbad

def get_star_name(ra, dec):
    """
    별의 적경, 적위를 이용하여 별의 이름을 찾는 함수

    Args:
      ra (str): 적경 (예: '05h34m31.95s')
      dec (str): 적위 (예: '+05d22m31.9s')

    Returns:
      str: 별의 이름 (없는 경우 None 반환)
    """

    # 좌표 객체 생성
    coord = SkyCoord(ra, dec, frame='icrs')

    # SIMBAD 쿼리
    result_table = Simbad.query_region(coord, radius = 5*u.arcsec)

    # 결과 출력 (첫 번째 결과만 반환)
    if len(result_table) > 0:
        return result_table['MAIN_ID'][0]
    else:
        return None
```

```python

# 예시 사용
star_name = get_star_name(targ_sky.ra, targ_sky.dec)

if star_name:
  print(f"적경 {targ_sky.ra}, 적위 {targ_sky.dec}에 해당하는 별의 이름은 {star_name}입니다.")
else:
  print("해당 좌표에 해당하는 별을 찾을 수 없습니다.")
```
