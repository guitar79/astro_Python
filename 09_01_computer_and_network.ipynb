### 슬라이딩 망원경 제어 컴퓨터

## NAS

NAS를 사용하고 싶은 경우 지구과학 선생님께 계정을 받아서 사용할 수 있습니다.

 ### GSESCloud

* Synology NAS
* https://quickconnect.to/GSESCloud 로 접속 가능
* ASTRO_data 폴더에 관측 데이터 저장하고 있음


### GSASTROCloud

* Synology NAS
* https://quickconnect.to/GSASTROCloud 로 접속 가능
* 관측 자료 원본을 저장하고 있음.
