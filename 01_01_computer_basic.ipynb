# 컴퓨터 기초

최근의 데이터들은 대부분 컴퓨터 파일 형태의 디지털로 제공이 되고 있고, 데이터 사이언스를 시작하기 위해서는 부득이 컴퓨터 기초에 대해 정확하게 이해를 할 필요가 있습니다. 최근 컴퓨터를 사용은 매우 친숙하지만 의외로 기초적인 내용을 잘 모르는 경우들이 있어서 잘 아는 사람들에게는 너무 당연한 내용이겠지만 기초적인 내용을 하나씩 짚고 넘어가는 것이 좋을 것으로 생각되어 컴퓨터 기초를 간단하게 다루고 넘어가고자 합니다. 

## 운영체제 

컴퓨터를 사용하기 위해서는 여러 요소들이 필요한데 그 중에 중요한 것 중에 하나가 바로 운영체제(OS, Operating System)입니다. 운영체제는 컴퓨터 시스템의 자원들을 효율적으로 관리하며, 사용자가 컴퓨터를 편리하고, 효과적으로 사용할 수 있도록 환경을 제공하는 여러 프로그램의 모임입니다. 운영체제는 컴퓨터 사용자와 컴퓨터 하드웨어 간의 인터페이스로서 동작하는 시스템 소프트웨어의 일종으로, 다른 응용프로그램이 유용한 작업을 할 수 있도록 환경을 제공합니다. 개인 컴퓨터(PC, Personal computer)에 많이 사용되는 운영 체제는 마이크로소프트의 윈도우(Microsoft windows), 애플의 macOS, 리눅스(Linux) 등이 있습니다. 

### 윈도우

컴퓨터에 관심이 있는 사람들은 관련 서적에서 MS-DOS(Microsoft Disk Operating System)나 플로피 디스크(floppy disk)에 관한 얘기를 접해보았을 수 있습니다. MS-DOS는 마이크로소프트사가 IBM사의 의뢰를 받아 시애틀 컴퓨터시스템사로부터 사들여 개발한 IBM PC용 운영 체제로 개발되어 대중화된 운영 체제입니다. 애플이 개인용 컴퓨터에 처음으로 도입한 그래픽 사용자 인터페이스(GUI, Graphic User Interface) 운영 체제에 대항하여, 당시 널리 쓰이던 MS-DOS에서 멀티 태스킹과 GUI 환경을 제공하기 위한 응용 프로그램으로 마이크로소프트 윈도우(Microsoft Windows)가 출시되었습니다. 

윈도우 1.0부터 3.1x에 이르는 윈도우 초기 버전들은 운영 체제가 아닌, MS-DOS 하에서 동작하는 일종의 소프트웨어로 간주되었습니다. 배경화면도 없고 windows.exe같은 글자와 물론 시작 버튼도 없고 명령어를 입력했기 때문에 사람들이 쓰기 불편했습니다. 이후 윈도우는 꾸준히 발전되어 현재는 멀티 유저, 먼티 테스킹이 지원되는 운영체제로 현재 전 세계 90%의 개인용 컴퓨터에서 쓰고 있으며, 서버용 운영 체제로도 점차 영역을 넓혀 나가고 있습니다.

### macOS

macOS(맥 오에스, 이전 명칭: OS X, 맥 OS X / Mac OS X)는 애플이 개발한 유닉스 기반 운영 체제입니다. 최초의 유닉스인 AT&T 벨연구소 유닉스의 소스코드를 이어받은 유전적 유닉스에 속하며 동시에 유닉스 상표권을 소유하며 유닉스 표준을 만든 오픈그룹으로부터 유닉스 인증을 받아 기술적으로도 법률적으로도 완벽하게 유닉스 기준을 충족합니다. 스티브 잡스가 설립한 회사 NeXT에서 만든 NeXTSTEP 운영체제의 후속작인 Darwin 운영체제의 후속작입니다. macOS의 최초의 ARM 아키텍처 운영체제가 iOS이며, 스티브 잡스가 "Mac OS X이 모바일로 돌아간다!"라고 홍보를 하였습니다.
사용자 입장에서는 한국에서는 ActiveX, 업무 관리, NEIS 등 특수한 환경 때문에 사용상 애로 사항이 많다. 이는 Windows가 아닌 모든 운영체제 사용자가 공통으로 겪게 되는 문제라 할 수 있으며 심지어 최신 버전의 Windows도 겪게 된다. 다행히 2010년 이후로 점점 사정이 나아지고 있어 어느 정도 숨통이 트이기 시작했으나 여전히 애로사항이 많습니다. 하지만 연구 활동(지도), 코딩 등의 작업에서는 유리한 면이 많이 있어 사용해 보고 싶은 운영 체제 입니다.

### 리눅스

컴퓨터 OS 커널의 일종인 리눅스 커널, 또는 리눅스 커널을 사용하는 운영체제를 가리키는 말이기도 합니다. GNU 쪽 사람들은 리눅스는 커널일 뿐이고, 이 커널을 가져다가 GNU 프로그램들을 올려 만든 운영체제는 GNU/Linux라고 이야기하며 이런 명칭에 민감하게 반응하는 경우도 있습니다. 소스 코드가 공개되어 있는 대표적인 오픈 소스 소프트웨어 입니다. 컴퓨터 역사상 가장 많은 참여자가 관여하고 있는 오픈 소스 프로젝트다. 모바일 운영체제로 유명한 안드로이드 역시 리눅스 커널을 가져다 쓰고 있습니다.

리눅스 OS 커널 소스코드는 GPLv2 라이선스입니다. Linux 등록 상표 사용에 대해서는 2008년 까지 유료였으나 그 이후 부터 무료로 사용이 가능하다. 현재는 우분투(Ubuntu), CentOS 등의 배포판이 있습니다.

## 콘솔

원래 콘솔(console)은 입력 장치와 출력 장치를 총칭하는 말로 컴퓨터의 대표적인 콘솔은 키보드와 모니터라고 볼 수 있습니다. 하지만 최근에는 콘솔 창이라 하면, 모니터를 뜻하기 보다는 터미널 창을 의미하는 것이 일반적입니다. 최근의 OS들은 GUI(Graphical User Interface)가 잘 만들어져 있어서 마우스로 화면을 클릭하면서 컴퓨터를 사용하지만 과거에는 콘솔 창을 통새서 단문의 명령어를 실행하는 형태로 컴퓨터를 사용하였습니다. 우리가 코딩을 통해 데이터를 처리할때 콘솔 창을 자주 사용하게 될 것이므로 OS 별로 콘솔 창을 여는 방법을 알 필요가 있습니다.   

### 윈도우

윈도우 사용자라면 다음과 같은 방법으로 기본적인 콘솔창을 열 수 있습니다. 

```{admonition} windows
* 시작 > cmd [enter]
```

cmd.png

윈도우에도 다른 파워쉘과 같이 다른 터미널이 있습니다.

```{admonition} windows
* 시작 > windows posershell [enter]
```

### macOS와 linux

macOS와 linux는 모두 UNIX에 뿌리를 두고 있기 때문에 멀티 유저, 멀티 테스킹이 기본이며, 여러 개의 터미널 창을 열고 동시에 여러 작업을 명령할 수 있습니다.

우분투의 경우에는 다음의 단축키로 터미널 창을 열 수 있습니다.
```{admonition} linux
* [ctrl] + [alt] + [T]
```

Screenshot from 2023-03-29 09-05-00.png

Screenshot from 2023-03-29 09-01-44.png

macOS 와 linux는 멀티 유저, 멀티 태스킹으로 설계된 UNIX에 기반을 두고 있어 콘솔 명령어는 거의 같게 사용할 수 있고, 파일과 디렉토리 모두 사용자와 그룹의 퍼미션을 각각 설정할 수 있습니다. 

|         명령어         |                                                                      설명                                                                     |
|:----------------------:|:---------------------------------------------------------------------------------------------------------------------------------------------:|
| ls              | 현재 디렉토리의 파일 목록을 출력합니다.                                                                                                      |
| cd                 | 디렉토리를 이동합니다.                                                                                                       |
| pwd                    | 현재 경로를 보여준다.                                                                                                                                |
| rm              | 파일이나 디렉토리를 삭제합니다.                                                                                                      |
| cp                 | 파일이나 디렉토리를 복사합니다.                                                                                                       |
| touch                    | 새로운 파일을 생성하거나, 이미 파일이 존재합니다면 최종 수정 시간을 현재 시간으로 변경합니다.                                                                                                                     |
| mv              | 이름을 변경하거나 다른 디렉토리로 이동시킵니다.                                                                                          |
| mkdir                | 새로운 디렉토리를 생성합니다.                                                                                      |
| rmdir                   | 디렉토리를 삭제합니다.                                                                                                            |
| cat              | 파일 내용을 보여 줍니다. 여러 파일의 내용을 연결해서 볼 수도 있습니다.                                                                                         |
| head, tail             | 파일의 앞 10행 또는 끝 10행을 보여 줍니다.                                                                                    |
| more, less             | 텍스트 형식의 파일을 페이지 단위로 보여 줍니다.                         |
| file             | 파일의 종류를 보여 줍니다.                     |
| clear             | 터미널 화면을 정리합니다.                |


## 파일과 폴더

### 파일 

파일(file)은 데이터 모음으로 구성되며 OS는 보조기억장치의 한 종류로 저장 장치인 하드 디스크 드라이브(HDD, Hard Disk Drive), 솔리드 디스크 드라이브(SSD, Solid State Drive) 등에 파일을 저장할 수 있습니다. 사용자는 파일 열기, 읽기, 수정 및 닫기와 같은 작업을 수행 할 수 있습니다. 또한 일부 파일은 폴더로 압축되지만 압축한 ZIP 파일은 많은 관련 파일과 폴더가있는 폴더이지만 단일 파일로 간주됩니다. 다양한 유형의 파일이 있으며 데이터 사이언스에서 파일을 다루는 방법을 잘 알 필요가 있습니다. 파일의 종류는 파일명의 확장자로 구분하는 것이 일반적이며 대표적인 파일의 확장자는 다음 표와 같다. 한 형식에서 다른 형식으로 파일을 변환 할 수도 있는데, 예를 들면 doc 형식 파일을 pdf 파일로 변환 할 수 있습니다.

#### 텍스트 파일

글자, 문단 등의 속성은 없고, 문자만으로 이루어진 파일입니다. 대부분 코딩은 텍스트 파일로 이루어져 있으며, 문자셋에 따라 인코딩을 신경써야 합니다.

| text 파일  |                 |
|-----------|------------------|
| .txt      | 일반 텍스트        |
| .c      | C 언어              |
| .py      |  Python           |
| .html      | 웹 페이지         |
| .script      | 스크립트 파일    |
| .bat      | batch 파일        |
| .csv      | omma-separated values        |



#### 문서 파일

글자, 문단 등의 속성이 있고, 그림, 표 등을 삽입할 수 있는 파일로 스프레드 시트와 같이 수식 계산 기능에 특화된 경우도 있습니다. 

| 문서 파일  |  |
|-----------|------------------|
| .hwp      | 한글과 컴퓨터     |
| .doc      | MS Word         |
| .ppt      | MS Power Point   |
| .pdf      |  Portable Document Format   |
| .xls      | MS Excel         |

#### 이미지 파일

이미지 파일도 벡터 이미지, 레스스터 이미지로 나눌 수 있고, 각각의 이미지도 색 심도, 압축 형태 등에 따라서 여러 포멧이 있습니다.

| 이미지파일 |                          |
|------------|--------------------------|
| .gif       | 움짤 등                  |
| .jpg       |                          |
| .png       | 압축,                    |
| .bmp       |                          |
| .tiff      |                          |
| .raw       |                          |

#### 동영상 파일

대부분의 동영상 파일은 용량이 커서 코덱을 사용하여 인코딩/디코딩 하는 기술이 매우 중요하다. 

| 동영상파일  |                                 |
|-----------|---------------------------------|
| .mov      | Media Player 외 (곰플레이어 외) |
| .mkl      |                                 |
| .fiv      |                                 |
| .avi      |                                 |
| .mp4      |                                 |

#### 압축 파일

폴더를 포함한 파일을 압축하여 저장하는 형식으로 최근에는 통합 압축 프로그램 이 발달하여 쉽게 압축을 해제할 수 있습니다.

| 압축파일  |                  |
|-----------|------------------|
| .zip      |                  |
| .alz      |                  |
| .egg      |                  |
| .rar      |                  |
| .tar      |                  |
| .gz       |                  |



#### 오디오 파일

음성, 음악 등의 소리를 저장하는 파일 형식들은 다음과 같습니다.

| 오디오파일 |                            |
|------------|----------------------------|
| .wav       |                            |
| .wma      |                            |
| .mp2       |                            |
| mp3        |                            |



#### 기타 

응용 프로그램들은 각각의 목적에 맞게 프로젝트를 생성하여 필요한 속성들을 저장하고 있습니다.

| 프로젝트 파일 |                    |
|------------|--------------------|
| .psd       | Adobe Photoshop   |
| .ai        | Adobe Illustrator |


### 폴더

폴더(folder)는 파일과 폴더를 저장하는데 디렉토리(directory)라고도 합니다. 모든 파일이 같은 위치에 있으면 특정 파일을 식별하기가 어려울 수 있리 때문에 폴더는 파일을 정렬하는 데 유용하다. 예를 들어 사용자는 기록 된 문서를 하나의 폴더에 저장하고 오디오 파일을 다른 폴더에 저장하는 방법으로 파일을 구성하고 분리할 수 있습니다. 또한 폴더는 모든 형식의 여러 파일로 담아둘 수 있고 자체 폴더에 다른 폴더를 저장할 수도 있습니다. 

## 경로

코딩을 하거나 데이터 처리를 하다보면 파일과 폴더의 경로를 정확하게 이해하는 것이 필요하다. 어떤 경우에는 절대 경로로 처리를 하기도 하고 어떤 경우에는 상대 경로 로 처리를 하기도 하는데 절대 경로와 상대 경로의 개념이 확실하지 않으면 생각보다 혼란을 격게 된다. 또한 OS 사이에 표현하는 방식도 약간의 차이가 있어 자신이 사용하는 OS에서 경로를 어떻게 표현하는지 정확하게 히해하고 있어야 합니다.

#### 절대 경로

컴퓨터의 파일을 찾아가는 방법은 절대 경로로 파일을 찾는 방법 한가지 뿐입니다. 절대 경로란 최초의 시작점으로 경유한 경로를 전부 기입하는 방식입니다. 어떠한 OS던 이렇게 최상위 루트 로부터 경유한 경로를 전부 기입한 절대 경로로 파일을 찾을 수 있습니다.

* 윈도우 OS의 바탕화면에 위치한 test.txt 파일을 예로 들자면 test.txt파일의 절대 경로는 일반적으로 C 드라이브의 최상위 루트 디렉토리로부터 시작되어 아래와 같이 구성된다.

```{admonition} windows
* C:\Users\UserID\Desktop\test.txt
```

* linux OS의 바탕화면에 위치한 test.txt 파일의 절대 경로는 일반적으로 최상위 루트 디렉토리로부터 시작되어 아래와 같이 구성된다.

```{admonition} linux
* /home/UserID/Desktop/test.txt
```

* mac OS의 바탕화면에 위치한 test.txt 파일의 절대 경로는 일반적으로 최상위 루트 디렉토리로부터 시작되어 아래와 같이 구성된다.

```{admonition} linux
* /home/UserID/Desktop/test.txt
```

### 상대 경로

상대 경로는 상대적 이라는 단어에서 알 수 있습니다시피 항상 비교할 대상이 있어야 합니다. 우리가 상대 경로로 파일을 읽을 때도 어떤 경로로 부터 비교하는지를 염두해야 합니다. 예를 들어 아래의 두 상대 경로가 있습니다고 해 보자.

```{admonition} windows
* Desktop\test.txt
* test.txt
```
```{admonition} linux
* Desktop/test.txt
* test.txt
```

OS는 절대 경로로 파일을 읽는 방법 밖에 없기 때문에 저렇게 경로를 알려준다면 OS는 파일의 위치를 찾지 못합니다. 위의 2개의 상대 경로는 비교 대상이 누구인지에 따라 경로가 바뀌게 된다.

먼저 윈도우즈의 경우 비교 대상이 만약, C 드라이브의 Parks 디렉토리라면, 위의 2개의 상대 경로는 아래와 같이 절대 경로로 해석된다.

```{admonition} windows
* C:\Parks\Desktop\test.txt
* C:\Parks\test.txt
```

이번엔 리눅스의 경우 비교 대상이 만약, /usr/local 디렉토리라면, 위의 2개의 상대 경로는 아래와 같이 절대 경로로 해석된다.
```{admonition} linux
* /usr/local/Desktop/test.txt
* /usr/local/test.txt
```

상대 경로는 기준 경로를 기준으로 절대 경로가 구성되며 상대 경로로 파일의 위치를 찾을 수 있는 이유는 기준 경로가 절대 경로로 변환하여 OS에게 전달 하기 때문입니다. 

절대 경로는 정적인 문자열로 특정 컴퓨터의 파일 위치를 정확히 알려주지만 경로를 다룰때는 이러한 정적인 특징이 오히려 단점이 될 수도 있습니다. test.txt의 경로가 자주 바뀌는 상황이거나 루트 디렉토리가 서로 다른 OS들을 사용합니다면 절대 경로의 정적인 특징 때문에 절대 경로로 작성된 모든 문서를 다시 작성하거나 OS 별로 절대 경로를 작성해서 관리해야 하는 불편함이 있습니다. 만약 상대 경로를 이용하여 비교 대상이 우리에게 주어지는 상황이라면 위에서 언급한 문제를 해결 할 수 있을 것입니다.
